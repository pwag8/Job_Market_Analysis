```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.sample(5)
```

```python
df.index
```

```python
df.index.name = 'job_index' # assign a name for the row index column
```

```python
df.head() # see the new name appearing over the index column
```

```python
df.index.dtype # check the data type (dtype) of this index field (it's an integer)
```

```python
# let's create a new df filtered for jobs in the US
df_usa = df[df['job_country'] == 'United States']
df_usa # notice how our index is no longer stepping by 1 every line, because we filtered and removed a bunch of rows. let's reset the index...
```

```python
df_usa.reset_index(inplace=True)

df_usa # notice that it keeps the original job_index column, but creates a new index for this df
# could be useful to leave this in case later you want to merge dataframes and use the original index as a matching key
```

```python
# you can SET the index using an existing column (the existing column becomes your new index)
# let's set the index back to the original index, the job_index column, as an example
df_usa.set_index('job_index', inplace=True)

df_usa # this actually drops the new unnamed index column we had added before
```

```python
# you can also SORT by an index

# let's use this pivot table as an example
median_pivot = df_usa.pivot_table(values='salary_year_avg', index='job_title_short', aggfunc=['median', 'min', 'max'])

median_pivot # see that the job_title_short is the index of this table
```

```python
median_pivot.sort_index(inplace=True) # sorts the index (job_title_short) alphabetically

median_pivot
```

```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_DA = df[df['job_title_short'] == 'Data Analyst'].copy() # create a new dataframe by filtering for only data analyst. to work with as an example

df_DA
```

```python
# let's aggregate by month. we need to extract the month out of the job_posted_date column

df_DA['job_posted_month_no'] = df_DA['job_posted_date'].dt.month

df_DA
```

```python
df_DA_explode = df_DA.explode('job_skills')

df_DA_explode
```

```python
# next we need to make a pivot table to aggregate skills by month for easy plotting

df_DA_pivot = df_DA_explode.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0)

df_DA_pivot

# the fill_value argument tells it to convert NaN values to zero, which is what we need for plotting
```

```python
# there are so many skills, too many to plot. let's total them up, then we can plot the most common skills only

df_DA_pivot.loc['Total'] = df_DA_pivot.sum() # add a total row

df_DA_pivot
```

```python
df_DA_pivot.loc['Total'].sort_values(ascending=False)

# we need to extract those top job_skills...
```

```python
df_DA_pivot[df_DA_pivot.loc['Total'].sort_values(ascending=False).index]
```

```python
df_DA_pivot = df_DA_pivot[df_DA_pivot.loc['Total'].sort_values(ascending=False).index]

df_DA_pivot = df_DA_pivot.drop('Total')

df_DA_pivot
```

```python
df_DA_pivot.iloc[:, :5]
```

```python
df_DA_pivot.iloc[:, :5].plot(kind='line')

plt.title('Top 5 Skills for Data Analysts per Month')
plt.ylabel('Count')
plt.xlabel('')
plt.show()
```

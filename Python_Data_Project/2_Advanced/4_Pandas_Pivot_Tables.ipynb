```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.pivot_table(index='job_title_short', aggfunc='size')
```

```python
df.pivot_table(values='salary_year_avg', index='job_title_short', aggfunc='median')
```

```python
df.pivot_table(values='salary_year_avg', index='job_title_short', columns='job_country', aggfunc='median')
```

```python
df.pivot_table(values='salary_year_avg', index='job_country', columns='job_title_short', aggfunc='median')
```

```python
top_countries = df['job_country'].value_counts().head(6) # let's find the most plentiful countries in the dataframe to make a chart with them

top_countries
```

```python
top_countries = df['job_country'].value_counts().head(6).index # I really just care about the index of those top 6 countries...

top_countries
```

```python
df_job_country_salary = df.pivot_table(
    values='salary_year_avg',
    index='job_country',
    columns='job_title_short',
    aggfunc='median'
)

df_job_country_salary = df_job_country_salary.loc[top_countries] # let's establish a new df that filters for only the countries in my top_countries list above

df_job_country_salary
```

```python
# now let's filter for only three specific job titles to make a good graph from it
job_titles = ['Data Analyst', 'Data Engineer', 'Data Scientist'] # just create a list of the job titles you want

df_job_country_salary = df_job_country_salary[job_titles]

df_job_country_salary
```

```python
# now let's plot it on a graph
df_job_country_salary.plot(kind='bar')
```

```python
# format the graph a bit

df_job_country_salary.plot(kind='bar')
plt.ylabel('Median Salary ($USD)')
plt.xlabel('')
plt.title('Median Salary by Country and Job Title')
plt.xticks(rotation=45, ha='right')
plt.show()
```

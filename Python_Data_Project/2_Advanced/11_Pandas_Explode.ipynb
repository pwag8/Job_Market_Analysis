```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df[['job_title_short', 'job_skills']].head(5) # we are going to explode BY the job_skills column (a column of list values)
```

```python
# if you explode by a column that contains lists, it will create a df with individual rows for each list value

df.explode('job_skills')

# notive the job_skills column now only has one value per row, and the row index is repeated for each iteration of the original row
```

```python
df_exploded = df.explode('job_skills')

# now we count up the number of times each job_skill appears

df_exploded['job_skills'].value_counts().head(10).plot(kind='bar')
```

```python
# what if we wanted to group the job skills by job titles

df_exploded.groupby(['job_title_short', 'job_skills']).size()
```

```python
skills_count = df_exploded.groupby(['job_title_short', 'job_skills']).size()

type(skills_count) # it's a series, but we really would rather it be a dataframe. let's convert it
```

```python
df_skills_count = skills_count.reset_index(name='skill_count')

df_skills_count = df_skills_count.sort_values(by='skill_count', ascending=False)

df_skills_count
```

```python
job_title = 'Data Analyst'
top_skills = 10

df_skill_final = df_skills_count[df_skills_count['job_title_short'] == job_title].head(top_skills)

df_skill_final.plot(kind='barh', x='job_skills', y='skill_count')
```

```python
# let's invert the y axis so it looks a little better. there's a method for that

df_skill_final.plot(kind='barh', x='job_skills', y='skill_count')
plt.gca().invert_yaxis()
plt.title(f'Top {top_skills} Skills for {job_title}s') # this is a dynamic f strings that plugs in your assigned variables (because you might go back and change the variables)
plt.xlabel('Job Posting Count')
plt.ylabel('')
plt.legend().set_visible(False) # hide the legend
plt.show()
```
